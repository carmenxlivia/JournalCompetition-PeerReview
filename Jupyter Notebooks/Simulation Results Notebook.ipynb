### Agents strategies exploration

For each agent strategy attribute, we calculate the average value and the standard deviation over all the agents. 

```python
feature_mean_df = pd.read_csv("MEAN_ATTRIBUTES.csv") 
feature_mean_df.head(5)
```

```python
feature_std_df = pd.read_csv("STD_ATTRIBUTES.csv") 
feature_std_df.head(5)
```

### $p_{m}$ MANUSCRIPT STRATEGY ERROR BAR

For manuscript strategy $p_{m}$ we produce an error bar. This shows at what value the manucript strategy converged. The plot also displays the variability the mansucript strategy at certain timesteps.

```python
timesteps = [float(x)*100 for x in range(100)]

manuscriptStrategy_mean_slice = feature_mean_df['manuscriptStrategy'][::10]
manuscriptStrategy_std_slice = feature_std_df['manuscriptStrategy'][::10]

plt.figure(figsize=(8,5))

plt.errorbar(timesteps,manuscriptStrategy_mean_slice,manuscriptStrategy_std_slice)
plt.title('$p_{m}$ manuscript strategy error bar', fontsize=20)

plt.xlabel('Time Step', fontsize=18)
plt.ylabel('Average value', fontsize=18)
```

### $Prop_{Q}$ and $Prop_{A}$ ERROR BARS

For both $Prop_{Q}$ and $Prop_{A}$, we produce an error bar. This shows how each attribute converges at certain values. The plot also displays the variability of the attributes at certain timesteps.

```python
timesteps = [float(x)*100 for x in range(100)]

propQ_mean_slice = feature_mean_df['propQ'][::10]
propQ_std_slice = feature_std_df['propQ'][::10]

propA_mean_slice = feature_mean_df['propA'][::10]
propA_std_slice = feature_std_df['propA'][::10]

fig, axs = plt.subplots(nrows=1, ncols=2, sharex=True,constrained_layout=True,  figsize = (15, 5))

axs[0].errorbar(timesteps,propQ_mean_slice,propQ_std_slice, label = 'propQ')
axs[1].errorbar(timesteps,propA_mean_slice,propA_std_slice, label = 'propA')

axs[0].title.set_text('$Prop_{Q}$ errorbar plot')
axs[1].title.set_text('$Prop_{A}$ errorbar plot')

axs[0].title.set_size(20)
axs[1].title.set_size(20)

axs[0].set_xlabel('Time step', fontsize = 15)
axs[0].set_ylabel('Average value', fontsize = 15)

axs[1].set_xlabel('Time step', fontsize = 15)
axs[1].set_ylabel('Average value', fontsize = 15)
```

### Journals' proportions

```python
proportions = pd.read_csv("PROPORTIONS.csv")  
proportions.head()
```

```python
n = 10
range_plot = range(len(proportions['Proportion Reviews'][::n]))
xlist = [x*100 for x in range_plot]

plt.figure(figsize=(8,5))

plt.plot(xlist,proportions['Proportion Papers'][::n], label = 'Good Papers over total')
plt.plot(xlist,proportions['Proportion Reviews'][::n], label = 'Good Reviews over total')

plt.xlabel('Time Step', fontsize=18)
plt.xticks(fontsize=14)

plt.ylabel('Proportion', fontsize=18)
plt.ylim(0.3, 1.1)
plt.yticks(fontsize=14)

plt.legend(loc='lower right',  prop={'size': 18})
plt.title('Evolution of the proportion of good papers and \n good reviews recorded in one simulation', fontsize=20)
```
